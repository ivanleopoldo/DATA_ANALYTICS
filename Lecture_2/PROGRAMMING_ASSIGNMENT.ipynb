```python
from ucimlrepo import fetch_ucirepo

dataset = fetch_ucirepo(id=374)

X = dataset.data.features
y = dataset.data.targets
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame(X)
dft = pd.DataFrame(y)

plt.figure(figsize=(12,6))
plt.plot(dataset.data.features['date'], dataset.data.targets['Appliances'])
plt.xlabel('Date')
plt.ylabel('Energy Use (Wh)')
plt.title('Appliances Energy Consumption Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['T1'], dft['Appliances'])
plt.xlabel('Temperature (C)')
plt.ylabel('Energy Use (Wh)')
plt.title('Correlation Between Temperature and Energy Use')
plt.show()
```

```python
room_energy = df[['lights', 'T1', 'T2', 'T3', 'T4', 'T5', 'T6', 'T7', 'T8', 'T9']].mean()
room_energy.plot(kind='bar', figsize=(10,6))
plt.xlabel('Room')
plt.ylabel('Average Energy Use (Wh)')
plt.title('Average Energy Use by Room')
plt.xticks(rotation=45)
plt.show()
```
